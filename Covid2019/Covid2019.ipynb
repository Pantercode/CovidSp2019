# Renomeado colunas

```python
def rename_column(df, old_column_name, new_column_name):
    df.rename(columns={old_column_name: new_column_name}, inplace=True)
```

```python
rename_column(df_covid, 'nome_munic', 'Municipio')
rename_column(df_covid, 'nome_drs', 'Cidade')
rename_column(df_covid, 'pop', 'População_abaixo_60')
rename_column(df_covid, 'pop_60', 'População_acima_60')
rename_column(df_covid, 'semana_epidem', 'Semana_epidemiologica')
```

# Exclusão de linhas

```python
def drop_column(df, column_name):
    if column_name in df.columns:
        df.drop(columns=[column_name], inplace=True)
    else:
        print(f"A coluna '{column_name}' não está inclusa no Dataframe.")
```

```python
drop_column(df_covid,'cod_ra')
drop_column(df_covid,'rotulo_mapa')
drop_column(df_covid,'cod_drs')
drop_column(df_covid,'nome_ra')
```

# Exclusão de linhas

```python
def drop_column_by_index(df, column_index):
    df.drop(df.columns[column_index], axis=1, inplace=True)
```

# Tratando dados Nulos e NAN

```python
def replace_empty_values(df, col, texto):
    df.loc[(df[col].isna()) | (df[col] == '') | (df[col] == ' '), col] = texto
```

```python
replace_empty_values(df_covid,'Cidade','Ignorado')
replace_empty_values(df_covid,'map_leg','Ignorado')
replace_empty_values(df_covid,'map_leg_s','Ignorado')
replace_empty_values(df_covid,'densidade','Ignorado')
df_covid
```
